```python
import pandas as pd
```

```python
first_half_comments = pd.read_csv('ebc_comments_3480.csv', header=0)
second_half_comments = pd.read_csv('ebc_comments_3481-end.csv', header=0)

first_half_article_links = pd.read_csv('ebc_article_links_3480.csv', header=0)
second_half_article_links = pd.read_csv('ebc_article_links_3481-end.csv', header=0)
```

```python
frames_comments = [first_half_comments, second_half_comments]
frames_article_links = [first_half_article_links, second_half_article_links]
```

```python
comments = pd.concat(frames_comments)
article_links = pd.concat(frames_article_links)
```

```python
comments.to_csv('ebc_comments.csv', index=False)
article_links.to_csv('ebc_article_links.csv', index=False)
```
